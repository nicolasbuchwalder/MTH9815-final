```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

```python
years = np.array([2, 3, 5, 7, 10, 30])
PV01s =[0.019851, 0.029309, 0.048643, 0.065843, 0.087939, 0.184698]
```

```python
plt.plot(years, PV01s)
```

```python
cs = CubicSpline(years, PV01s)
```

```python
# cubic spline interpolation
cs(20)
```
